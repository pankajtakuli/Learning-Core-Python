- To create an array: ***np.array([elements])***
- To access an individual elemnet: ***array[index_value]***
- To find numbers of dimensions: ***array.ndim***
- To find the shape of an array: ***array.shape***
- To find the size of an array: ***array.size***
- To find the data type of an element: ***array.dtype***


<Font size = '5'>***We can create an array by some others methods-***

- By zeros method: ***np.zeros()***
- By ones method: ***np.ones()***
- By empty method: ***np.empty***
- By arange(start, stop, step): ***np.arange(start, stop, step)***
- By linspace(start, stop, step): ***np.linspace(start, stop, step)***

- Sorting: ***np.sort()***
- Concatenation: ***np.concatenate((array1, array2))***
- To add two array horizontally we use ***np.concatenate((array1 , array2), axis = 0)***
- To add two array vertically we use ***np.concatenate((array1 , array2), axis = 1)***

- To shape array without changing data of array: ***array.reshape()***


***Indexing using conditiions of Array***
- a[condition]

***Basic operatioins of Array***
- Sum +
- Subtraction -
- Multiplication *
- Division /

- To add all the elements of the array: ***array.sum()***
