## What is Data?

Data refers to information or observations that are collected, recorded, or analyzed to draw conclusions, make interfences, or support decision-making.

### _Types of Data:_

***1.Categorical Data (Classified Data)***
- Categorical data represents groups or categories.
- ***Examples-***
1. Car brands : Audi, BMW, and Mercedes.
2. Answers to yes/no questions : yes and no


***2.Numerical Data***
- Numerical data represents numbers.
- ***Examples-***
1. Marks of a students in English.
2. Sales of a company.
3. Income of employees of a company.
4. Height of each student in a class.

### Numerical Data

***1.Discrete data***
- Discrete data take on specific, distinct values and are counted in individual units. The values are often integers, and are seperate from each other.
- ***Eg.:***
1. Number of children in a school.
2. Number of books in your libraray.
3. Number of cases a Lawayer has won.

***2. Continuous data***
- Continuous data can take on any value within a given range.
- ***Eg:***
1. Height of an individual 175.78 , 150.05cm
2. Monthly Expenditure of a company. Rs.120000034.34 , Rs.23432.34

***Nominal data***
- The nominal level represents categories that cannot be put in any order.
- Eg:
  1. Car brands : Maruti, Honda , Mahindra, Hyundai
  2. Subjects : Maths, Humanities , Commerce , Biology


***Ordinal data***
- Eg:
  1. Rating of a food - Neutral , Tasty
  2. Programing level - Beginner , Intermediate , Advanced

***Interval data***
- This level of measurement has no true zero. This means that a zero does not mean absence of a property.
- Eg:
  1. Temperature in Celsius or Fahrenheit : 0c or 0F does not mean the lowest temperature.
  2. Credit scores range from 200-800. Here 0 does not mean true zero.


***Ratio data***
- This level of measurement has a true zero.
- Eg:
  1. Temperature in Kelvin: 0k means the lowest possible temperature
  2. Height of an object - if the height of an object is 0cm then it means no height

<font size = '5'>***Measures of Central Tendancy***

- Central tendancy is a way to describe the "center" of "typical" value of a set of numbers. It gives you a sense of where most of the data is concntrated.
- You can "make sense" of the data without looking at all the values or observations by just looking at the central tendency.
- A measure of central tendancy is a summary measure of your column.

- There have been several measures of central tendency developed in statistical theory. The most commonly used ones include:
  - Mean or Arithmetic mean
  - Median
  - Mode

<font size = '5'>***Mean***
- It can be calculated as - <b>Sum of all the values / Numbers of values

<font size = '5'>***Median***

- It can be calculated by arranging values in either ascending or descending order and then finding the middle value.
- For odd number of values:
  - <b>_Median = ((n + 1)/2)th term_<b>
- For even numbers of values:
  - <b>_Median = ((n/2)th term + (n/2 + 1)th term)/2_

<font size = '5'>***Mode***

- The mode of the data is the most frequently occuring value.
- The mode can be used for nominal data (Categorical Data).

<font size = '5'><b>***Measures of Dispersion***
- The measures of dispersion is an index which quantifies the degree to which the data is dispersed or spread.
- It serves as the measures of the reliabilty of the average.

<font size = '5'>***Uses of Measures of Dispersion***
- It shows the variation in the data that provide information such as how well the average of the sample represent the entire data. Less variation gives close representation while with larger variation, the average may not closely represent all the values in the sample.
- It enalbles us to compare two or more series with regard to their varitions. It helps to determine consistency.
- It also helps to control the cause behind the variations.

- ***Measures of Dispersion***
  1. Range
  2. Interquartile Range
  3. Standard deviation
  4. Variance


<font size = '5'><b>Range
- Range is the difference between the largest and smallest data points in the set.
- Range = Max value - Min value
- For our example -
  - Player 1 = 5 , 15 , 15 , 25
  - Player 2 = 10 , 15 , 15 , 20
- Mean of Player 1 = 15
- Mean of Player 2 = 15
- Range of Player 1 = 25 - 5 = 20
- Range of Player 2 = 20 - 10 = 10
Thus, Player 2 is more consistent than player 1 and has more reliable average.

<font size = '5'>***Problems with the Range***
- ***Sensitive to Outlirs*** : The range is highly sensitive to extreme values (outliers) because it depends only on the maximum and minimum values. A single extreme value can significantly disort the measure.
  <b>Example :<b> Consider the following dataset: [10, 12, 15, 18, 20, 100]. The range is 100 - 10 = 90. However, the majority of values are within a small range , and the large outlier [100] greatly influences the result.

<font size = '5'>***Quartile***
- The first quarutile, represented as Q1, refers to the 25 percentile of the data.
- The second quartile , represented as Q2, refers to the 50 percentile of the data. It is actually the median of the data.
- The third quartile, represented as Q3, refers to the 75 percentile of the data.

<font size = '5'>***Interquartile Range(IQR)***
- Interquartile Range is the difference between the first (Q1) and the third quartile (Q3).
- It describes the middle 50% of the observations. If the middle 50% of the data points are spaced wide apart, their interquartile range will be large.
- ***IQR = Q3 - Q1***
- The IQR is primarily  used for detecting outliers. Values below ***Q1 - 1.5IQR*** and above ***Q3 + 1.5IQR*** are considered as outliers.
